```python
import heatextremes as he 
```

```python
era5 = he.open_cached_era5() 
threshold = 32
```

```python
days_exceeding_threshold = (era5['2m_temperature'] > threshold).groupby('time.year').sum()
```

```python
t2m_daily_mean = era5['2m_temperature'].resample(time="1D").mean()
t2m_daily_mean_doy_clim = t2m_daily_mean.groupby('time.day_of_year').mean()
```

```python
t2m_daily_max = era5['2m_temperature'].resample(time="1D").max()
t2m_daily_max_doy_clim = t2m_daily_max.groupby('time.day_of_year').mean()
```

```python
days_exceeding_threshold.mean('year').plot()
```
